```python
import matplotlib.pyplot as plt
import numpy as np

import os
import sys
```

```python
curdir = 'F:\Sem 5\Shewtank_BTP\\'

print(curdir)

# pic1 = 'bezier_gradient_2x.png'
pic1 = 'images.jpg'

content = os.listdir(curdir)
print(content)
```

```python
import matplotlib.image as mpimg 
from scipy.ndimage import gaussian_filter
```

```python
mat1 = mpimg.imread(curdir + pic1)

mat1 = gaussian_filter(mat1,0.5)

print(type(mat1))#.shape)
print('shape = ', mat1.shape)
plt.imshow(mat1) 
```

```python
dx = np.array([-1, 0, 1])
dy = np.array([-1, 0, 1])

m = mat1.shape[0]
n = mat1.shape[1]

fx = np.zeros([m,n])
fy = np.zeros([m,n])

# print(fx)
# print(fy)
```

```python
from skimage.color import rgb2gray
```

```python
# Calculating the gradient based on grayscale intensity

mat1_gray = rgb2gray(mat1)
plt.imshow(mat1_gray)

# print(mat1_gray[200][200])
```

```python
for i in range(1, m-1, int(m/10)):
    for j in range(1, n-1, int(n/10)):
        fx[i, j] = np.sum(mat1_gray[i-1:i+2, j] * dx)
        fy[i, j] = np.sum(mat1_gray[i, j-1:j+2] * dy)
```

```python
x = np.arange(0,mat1_gray.shape[0])
y = np.arange(0,mat1_gray.shape[1])

X, Y = np.meshgrid(x[0::int(m/10)], y[0::int(n/10)])

print(x[0::36].shape)
print(X.shape, Y.shape)

mag1 = np.sqrt(np.multiply(fx,fx) + np.multiply(fy,fy))

fx = fx/mag1
fy = fy/mag1

fig, ax = plt.subplots(figsize=(15,15))

q = ax.quiver(X,Y,fx,fy)
ax.quiverkey(q, X=0.3, Y=1.1, U=10,
             label='Quiver key, length = 10', labelpos='E')

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

plt.savefig('nice.png')

plt.show()
```
